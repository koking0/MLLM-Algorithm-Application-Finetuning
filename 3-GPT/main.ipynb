```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi
from numpy import tanh
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def gelu(x):
    return 0.5 * x * (1 + tanh(sqrt(2 / pi) * (x + 0.044715 * x ** 3)))
```

```python
# 创建x值
x = np.linspace(-3, 3, 1000)

# 计算ReLU和GeLU的y值
y_relu = relu(x)
y_gelu = gelu(x)

# 绘制ReLU和GeLU函数图像
plt.figure(figsize=(10, 6))
plt.plot(x, y_relu, color="red", label="ReLU")
plt.plot(x, y_gelu, color="green", label="GeLU")
plt.legend()
plt.grid(True)
plt.title("ReLU and GeLU functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义Softmax函数
def softmax(x):
    e_x = np.exp(x - np.max(x))  # for numerical stability
    return e_x / e_x.sum()

# 创建一个从-10到9的数组，并在末尾添加一个100
x1 = np.arange(-10, 10)
x1 = np.append(x1, 100)
# 计算Softmax值
y1 = softmax(x1)

# 创建另一个从-10到9的数组，在末尾添加一个100，然后除以100进行缩放
x2 = np.arange(-10, 10)
x2 = np.append(x2, 100)
x2 = x2 / 100
# 计算Softmax值
y2 = softmax(x2)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(x1)+1), y1, marker="o", color="blue", label="Original")
plt.plot(range(1, len(x2)+1), y2, marker="o", color="red", label="Scaled")
plt.xlabel("Index")
plt.ylabel("Softmax Value")
plt.title("Softmax Values of Two Vectors")
plt.grid(True)
plt.legend()
plt.show()
```
